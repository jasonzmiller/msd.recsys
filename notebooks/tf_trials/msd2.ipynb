```python
n_users = len(clean_df['user'].unique())
n_songs = len(clean_df['song'].unique())
print('There are {0} users and {1} songs in the data'.format(n_users, n_songs))
```

```python
# If we build a matrix of users x games, how many cells in the matrix will be filled?
sparsity = clean_df.shape[0] / float(n_users * n_songs)
print('{:.2%} of the user-item matrix is filled'.format(sparsity))
```

```python
# Create the dictionaries to convert user and songs to idx and back
user2idx = {user: i for i, user in enumerate(clean_df['user'].unique())}
idx2user = {i: user for user, i in user2idx.items()}

song2idx = {song: i for i, song in enumerate(clean_df['song'].unique())}
idx2song = {i: song for song, i in song2idx.items()}
```

```python
# Convert the user and games to idx
user_idx = clean_df['user'].apply(lambda x: user2idx[x]).values
song_idx = clean_df['songIdx'] = clean_df['song'].apply(lambda x: song2idx[x]).values
play_count = clean_df['count'].values
```

```python
# Using a sparse matrix will be more memory efficient and necessary for larger dataset, 
# but this works for now.

zero_matrix = np.zeros(shape = (n_users, n_songs)) # Create a zero matrix
user_song_pref = zero_matrix.copy()
user_song_pref[user_idx, song_idx] = 1 # Fill the matrix will preferences (bought)

user_song_interactions = zero_matrix.copy()
# Fill the confidence with (hours played)
# Added 1 to the hours played so that we have min. confidence for games bought but not played.
user_song_interactions[user_idx, song_idx] = play_count 
```

```python
k = 2

# Count the number of songs listened to by each user
purchase_counts = np.apply_along_axis(np.bincount, 1, user_song_pref.astype(int))
listeners_idx = np.where(purchase_counts[:, 1] >= 2 * k)[0] #find the users who listen to 2 * k songs
print('{0} users listened to {1} or more songs'.format(len(listeners_idx), 2 * k))
```

```python
test_frac = 0.2 # Let's save 10% of the data for validation and 10% for testing.
test_users_idx = np.random.choice(listeners_idx,
                                  size = int(np.ceil(len(listeners_idx) * test_frac)),
                                  replace = False)
```

```python
val_users_idx = test_users_idx[:int(len(test_users_idx) / 2)]
test_users_idx = test_users_idx[int(len(test_users_idx) / 2):]
```

```python
# A function used to mask the preferences data from training matrix
def data_process(dat, train, test, user_idx, k):
    for user in user_idx:
        purchases = np.where(dat[user, :] == 1)[0]
        mask = np.random.choice(purchases, size = k, replace = False)
        
        train[user, mask] = 0
        test[user, mask] = dat[user, mask]
    return train, test
```

```python
train_matrix = user_song_pref.copy()
test_matrix = zero_matrix.copy()
val_matrix = zero_matrix.copy()

# Mask the train matrix and create the validation and test matrices
train_matrix, val_matrix = data_process(user_song_pref, train_matrix, val_matrix, val_users_idx, k)
train_matrix, test_matrix = data_process(user_song_pref, train_matrix, test_matrix, test_users_idx, k)
```
